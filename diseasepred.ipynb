```python
pd.read_csv('gs://cloud-ai-platform-a6207a7e-8f4c-4719-adc1-b03ae03c7677/disease_pred_data.csv')
```

```python
dataset = aiplatform.TabularDataset.create(
    display_name="disease_pred_1",
    gcs_source=['gs://cloud-ai-platform-a6207a7e-8f4c-4719-adc1-b03ae03c7677/disease_pred_data.csv']
)
```

```python
job = aiplatform.AutoMLTabularTrainingJob(
    display_name="training_d_p",
    optimization_prediction_type="classification",
    column_transformations=[
        {"numeric": {"column_name": "Glucose"}},
        {"numeric": {"column_name": "Cholesterol"}},
        {"numeric": {"column_name": "Hemoglobin"}},
        {"numeric": {"column_name": "Platelets"}},
        {"numeric": {"column_name": "White_Blood_Cells"}},
        {"numeric": {"column_name": "Red_Blood_Cells"}},
        {"numeric": {"column_name": "Hematocrit"}},
        {"numeric": {"column_name": "Mean_Corpuscular_Volume"}},
        {"numeric": {"column_name": "Mean_Corpuscular_Hemoglobin"}},
        {"numeric": {"column_name": "Mean_Corpuscular_Hemoglobin_Concentration"}},
        {"numeric": {"column_name": "Insulin"}},
        {"numeric": {"column_name": "BMI"}},
        {"numeric": {"column_name": "Systolic_Blood_Pressure"}},
        {"numeric": {"column_name": "Diastolic_Blood_Pressure"}},
        {"numeric": {"column_name": "Triglycerides"}},
        {"numeric": {"column_name": "HbA1c"}},
        {"numeric": {"column_name": "LDL_Cholesterol"}},
        {"numeric": {"column_name": "HDL_Cholesterol"}},
        {"numeric": {"column_name": "ALT"}},
        {"numeric": {"column_name": "AST"}},
        {"numeric": {"column_name": "Heart_Rate"}},
        {"numeric": {"column_name": "Creatinine"}},
        {"numeric": {"column_name": "Troponin"}},
        {"numeric": {"column_name": "C_reactive_Protein"}}
    ]
)
```

```python

model = job.run(
    dataset=dataset,
    target_column="Disease",
    budget_milli_node_hours=1000,
    model_display_name="disease_pred_1",
    disable_early_stopping=False
)
```

```python
endpoint = model.deploy(machine_type='n1-standard-4')
```
